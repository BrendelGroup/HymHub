# HymHub: analysis demo

HymHub provides descriptive statistics of several genome features for 13 species in the order Hymenoptera. These statistics are stored in tabular format (tab-separated values or TSV) to facilitate loadeding into popular statistics packages for analysis. This notebook provides a brief demo for loading a data set into Python. This demo will focus on interval loci (iLoci), but data for several other genomic features is available.

## Read data into memory

The Pandas library provides a very simple interface for loading tabular data into a data frame.
We'll import this library and read the iLocus summary stats into memory.

```python
import pandas
loci = pandas.io.parsers.read_table("iloci.tsv")
loci.head(5) # Let's check out the first 5 rows of the data frame
```

## Get the number of iLoci for each species

iLocus data for all 13 species in HymHub are included in the same table.
The **Species** column is provided to help distinguish which species a particular iLocus is from (see [this page](http://brendelgroup.github.io/HymHub/genomes.html) for a full description).
We can grab the species values and determine how many rows (iLoci) correspond to each species.

```python
species = list(set(loci["Species"]))
for spec in sorted(species):
    count = sum(loci["Species"] == spec)
    print "%s: %d" % (spec, count)
```

## Plot distribution of iLocus GC content

Here, we discard the 10% most extreme iLoci in terms of length, and then plot the GC content of the remaining iLoci by species.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy
import scipy
import seaborn

# Discard extremely short and extremely long iLoci
lq = scipy.stats.mstats.mquantiles(loci["Length"], prob=[0.05, 0.95])
loci_s = loci.loc[(loci.Length > lq[0]) & (loci.Length < lq[1])]
lq = ["%.1lf" % x for x in lq]
print lq

# Configure plot
seaborn.set_style("white")
seaborn.set_context({"figure.figsize": (12, 5)})

# Plot histograms as lines
for spec in species:
    specgc = loci.loc[(loci.Species == spec), "GCContent"]
    y, bin_edges = numpy.histogram(specgc, bins=20)
    bin_centers = 0.5*(bin_edges[1:]+bin_edges[:-1])
    plt.plot(bin_centers, y, '-')
plt.show()
```
